```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
data = pd.read_csv('/home/sszu/Huawei/4-2汇总_去掉第一列.csv')

# 提取速度和TPI列
speed = data['平均速度']
tpi = data['TPI']/5

# 绘制散点图
# 绘制散点图
# 绘制散点图
# 绘制散点图
plt.figure(figsize=(10, 6))
plt.scatter(speed, tpi, alpha=0.5, color='lightblue', edgecolor='k', s=50)  # 调整点的颜色为淡蓝色
plt.title('Relationship between Speed and TPI', fontsize=16)
plt.xlabel('Average Speed', fontsize=14)
plt.ylabel('TPI', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns  # 导入seaborn库

# 绘制平均速度与TPI的关系（带密度）
plt.figure(figsize=(10, 6))
sns.kdeplot(x=speed, y=tpi, fill=True, cmap='Blues', alpha=0.3)  # 添加密度图
plt.scatter(speed, tpi, alpha=0.5, color='lightblue', edgecolor='k', s=50)
plt.title('Relationship between Average Speed and TPI with Density', fontsize=16)
plt.xlabel('Average Speed', fontsize=14)
plt.ylabel('TPI', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 绘制流量与TPI的关系（带密度）
plt.figure(figsize=(10, 6))
sns.kdeplot(x=data['流量'], y=tpi, fill=True, cmap='Reds', alpha=0.3)  # 添加密度图
plt.scatter(data['流量'], tpi, alpha=0.5, color='salmon', edgecolor='k', s=50)
plt.title('Relationship between Flow and TPI with Density', fontsize=16)
plt.xlabel('Flow', fontsize=14)
plt.ylabel('TPI', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.kdeplot(data['密度'], y=tpi, fill=True, cmap='Greens', alpha=0.5)  # 添加密度图
plt.title('Density of Density and TPI', fontsize=16)
plt.xlabel('Density', fontsize=14)
plt.ylabel('TPI', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 绘制密度与TPI的关系（带散点图）
plt.figure(figsize=(10, 6))
sns.kdeplot(data['密度'], y=tpi, fill=True, cmap='Greens', alpha=0.5)  # 添加密度图
plt.scatter(data['密度'], tpi, alpha=0.5, color='lightgreen', edgecolor='k', s=50)  # 添加散点图
plt.title('Relationship between Density and TPI', fontsize=16)
plt.xlabel('Density', fontsize=14)
plt.ylabel('TPI', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

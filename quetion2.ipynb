```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
data = pd.read_csv('/home/sszu/Huawei/4-2_.csv')

# TPI
speed = data['']
tpi = data['TPI']/5

# 
# 
# 
# 
plt.figure(figsize=(10, 6))
plt.scatter(speed, tpi, alpha=0.5, color='lightblue', edgecolor='k', s=50)  # 
plt.title('Relationship between Speed and TPI', fontsize=16)
plt.xlabel('Average Speed', fontsize=14)
plt.ylabel('TPI', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

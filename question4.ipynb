```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.seasonal import seasonal_decompose

# 读取数据
data = pd.read_csv('test/new_3-2汇总.csv')

# 提取“平均速度”列
average_speed = data['平均速度']

# 创建时间序列，从0开始
time = range(len(average_speed))

# 可视化平均速度随时间变化
plt.figure(figsize=(15, 6))
plt.plot(time, average_speed, label='Average Speed', color='blue')
plt.title('Average Speed Over Time')
plt.xlabel('Time (from 0)')
plt.ylabel('Average Speed')
plt.xlim(0, len(average_speed) - 1)  # 设置横轴范围从0开始
plt.legend(loc='upper left')
plt.grid(True)
plt.show()

# 进行季节性分解
decomposition = seasonal_decompose(average_speed, model='additive', period=12)  # 根据数据的周期设置
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# 可视化分解结果
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(time, average_speed, label='Original', color='blue')
plt.xlim(0, len(average_speed) - 1)
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(time, trend, label='Trend', color='orange')
plt.xlim(0, len(trend) - 1)
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(time, seasonal, label='Seasonal', color='green')
plt.xlim(0, len(seasonal) - 1)
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(time, residual, label='Residual', color='red')
plt.xlim(0, len(residual) - 1)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# 自相关图
plt.figure(figsize=(12, 6))
autocorrelation_plot(average_speed)
plt.title('Autocorrelation of Average Speed', fontsize=16, fontweight='bold')
plt.xlabel('Lag', fontsize=14)
plt.ylabel('Autocorrelation', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.seasonal import seasonal_decompose

# 设置 Seaborn 样式
sns.set(style="whitegrid")

# 读取数据
data = pd.read_csv('test/new_3-2汇总.csv')

# 提取“平均速度”列
average_speed = data['平均速度']

# 创建时间序列，从0开始
time = range(len(average_speed))

# 可视化平均速度随时间变化
plt.figure(figsize=(14, 7))
plt.plot(time, average_speed, label='Average Speed', color=sns.color_palette("deep")[0], linewidth=2)
plt.title('Average Speed Over Time', fontsize=18, fontweight='bold')
plt.xlabel('Time (from 0)', fontsize=14)
plt.ylabel('Average Speed', fontsize=14)
plt.xlim(0, len(average_speed) - 1)  # 设置横轴范围从0开始
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 进行季节性分解
decomposition = seasonal_decompose(average_speed, model='additive', period=12)  # 根据数据的周期设置
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# 可视化分解结果
plt.figure(figsize=(14, 10))
plt.subplot(411)
plt.plot(time, average_speed, label='Original', color=sns.color_palette("deep")[0], linewidth=2)
plt.xlim(0, len(average_speed) - 1)
plt.legend(loc='upper left', fontsize=12)
plt.title('Original Average Speed', fontsize=16)
plt.subplot(412)
plt.plot(time, trend, label='Trend', color=sns.color_palette("deep")[1], linewidth=2)
plt.xlim(0, len(trend) - 1)
plt.legend(loc='upper left', fontsize=12)
plt.title('Trend', fontsize=16)
plt.subplot(413)
plt.plot(time, seasonal, label='Seasonal', color=sns.color_palette("deep")[2], linewidth=2)
plt.xlim(0, len(seasonal) - 1)
plt.legend(loc='upper left', fontsize=12)
plt.title('Seasonal', fontsize=16)
plt.subplot(414)
plt.plot(time, residual, label='Residual', color=sns.color_palette("deep")[3], linewidth=1)
plt.xlim(0, len(residual) - 1)
plt.legend(loc='upper left', fontsize=12)
plt.title('Residual', fontsize=16)
plt.tight_layout()
plt.show()

# 自相关图，设置窄一点的宽度
plt.figure(figsize=(10, 7))  # 调整宽度
autocorrelation_plot(average_speed)
plt.title('Autocorrelation of Average Speed', fontsize=18, fontweight='bold')
plt.xlabel('Lag', fontsize=14)
plt.ylabel('Autocorrelation', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
